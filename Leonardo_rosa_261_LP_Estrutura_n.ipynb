### 3. Elabore uma estrutura para representar um produto (código, nome, preço). **Crie uma função** para cadastrar 5 produtos. **Crie outra função** para aplicar 10% de aumento no preço do produto e apresente, por meio **de outra função**, todos os dados do produtos cadastrados, após o aumento. Construa uma função para cada opção do menu a seguir:
Menu do Sistema
1. Cadastrar
2. Reajustar
3. Visualizar
4. Sair

Qual opção deseja?

```python
# CRUD - Create, Read, Update, Delete - Inserção, Visualização, Atualização, Exclusão
class TipoProduto: # tipo de dado heterogêneo (int, str, float)
    codigo = 0 # codigo é int
    nome = '' # nome é str
    preco = 0.0 # preco é float

def menu():
    print('\nMenu do Sistema')
    print('1 - Cadastrar')
    print('2 - Reajustar')
    print('3 - Visualizar')
    print('4 - alterar')
    print('5 - excluir')
    print('7 - Sair')
    return int(input('Qual opção deseja? '))

def cadastrar(vetp):
    for i in range(3):
        p = TipoProduto() #inicialização/preparação da variável p
        p.codigo = int(input('Digite o código: '))
        p.nome = input('Digite o nome: ')
        p.preco = float(input('Digite o preço: '))
        vetp.append(p)
    return vetp

def visualizar(vetp):
    if len(vetp) > 0: # se tiver produtos cadastrados mostre
        print('Código    Nome               Preço R$')
        for p in vetp:
            print(f'{p.codigo:10} {p.nome:20} {p.preco:>.2f}')
    else:
        print('Não há produtos cadastrados!')

def reajustar(vetp):
    if len(vetp) > 0: # se tiver produtos cadastrados mostre
        for p in vetp:
            p.preco = p.preco + (p.preco * 10 / 100)
        print('Reajuste realizado com sucesso!')
    else:
        print('Não há produtos cadastrados!')
    return vetp
def alterar(vetp):
    if len(vetp) > 0:
        codigo_pesquisa = int(input("Qual o código quer encontrar?"))
        for p in vetp:
            if codigo_pesquisa == p.codigo:
                p.nome = input("Digite o novo nome: ")
                p.preco = float(input("Digite o novo: "))
                break
    else:
        print("Você não tem produtos cadastrados!")
    return vetp

def excluir(vetp):
    if len(vetp) > 0:
        for in range(len(vetp)):
            if i == 0:
                menor_preco = vetp[i].preco
                indice = i
            if vetp[i].preco <= menor_preco:
                menor_preco = vetp[i].preco
                indice = i
        vetp.pop(indice)
        print("Exclusão realizada com sucesso")
    else:
        print("Você não tem produtos cadastrados!")
def main():
    vet_produto = [] # vetor vazio
    opcao = menu()
    while opcao >= 1 and opcao <= 6:
        if opcao == 1: # vet_produto entra vazio e
            # volta com conteúdo cadastrado
            vet_produto = cadastrar(vet_produto)
        elif opcao == 2:
            vet_produto = reajustar(vet_produto)
        elif opcao == 3:
            visualizar(vet_produto)
        elif opcao == 5:
            vet_produto = alterar(vet_produto)
        elif opcao == 6:
            vet_produto = excluir(vet_produto)
        opcao = menu() # última linha de código do while, faça o input novamente

main()
```

```python
class TipoAluno:
    nome = ''
    data= ''
    telefone = ''
    endereco = ''
    serie = ''

def menu():
    print('\nMenu do Sistema')
    print('1 - Cadastrar aluno')
    print('2 - Consulta por nome')
    print('3 - Visualizar todos os dados')
    print('4 - Sair')
    return int(input('Qual opção deseja? '))

def cadastrar(veta):
    print()
    print("### Cadastrar aluno ###")
    op = 0
    cont = 0
    while op = 0 and cont <= 500:
        print()
        a = TipoAluno()
        a.nome = input('Digite o nome do aluno: ')
        a.data = input('Digite a data de nascimento do aluno: ')
        a.telefone = input('Digite o telefone dos pais ou responsaveis: ')
        a.endereco = input('Digite o endereço onde o aluno mora: ')
        a.serie = input('Digite a série que o aluno está: ')
        veta.append(a)
        cont += 1

        chat = input("Deseja cadastrar um novo aluno [S] ou [N]?")
        chat = chat.upper()
        if chat == "S":
            op = 0
        if chat == "N":
            op = 1
        else:
            print('Opção Inválida!')
    return veta

def consultar(veta):
    if len(veta) > 0:
        nome_pesquisar = input('Qual nome quer pesquisar? ')
        achei = False
        for a in veta:
            if nome_pesquisar == a.nome:
                achei = True
                print('Nome:',a.nome,'Data de Nascimento:',a.data,'Telefone',a.telefone,'Endereço: ', a.endereco,'')
        if not achei:
            print('Aluno não encontrado!')
    else:
        print('Não há alunos cadastrados!')

def visualizar(veta):
     if len(veta) > 0:
        print(f"{'Código'}  {'Nome'}                  {'Preço R$'}")
        for a in veta:
            print(f'{a.codigo:<7} {a.nome:<15} {a.preco:>10.2f}')
    else:
        print('Não há produtos cadastrados!')

def main():
    vet_aluno = []
    opcao = menu()
    while opcao >= 1 and opcao <= 6:
        if opcao == 1:
            vet_aluno = cadastrar(vet_aluno)
        elif opcao == 2:
            consultar(vet_aluno)
        elif opcao == 3:
            visualizar(vet_aluno)
        opcao = menu()
main()
```

### 5. Faça um programa que realize o cadastro de contas bancarias com as seguintes informações: numero da conta, nome do titular e saldo. O banco permitirá o cadastramento de 15 contas. **Crie uma função para cada opção do menu** a seguir. Utilize a estrutura na passagem de parâmetro:

```python
class TipoPessoa: # tipo de dado heterogêneo (int, str, float)
    numero_conta = 0 # codigo é int
    nome = '' # nome é str
    saldo = 0.0 # preco é float

def menu():
    print('\nMenu do Sistema')
    print('1 - Cadastrar')
    print('2 - Visualizar todas as contas')
    print('3 - Consultar por nome')
    print('4 - Alterar nome e/ou saldo de um número de conta')
    print('5 - Excluir a conta com menor saldo')
    print('7 - Sair')
    return int(input('Qual opção deseja? '))

def cadastrar(vetp):
    for i in range(3):
        p = TipoPessoa()
        p.numero_conta = int(input('Digite o código: '))
        p.nome = input('Digite o nome: ')
        p.saldo = float(input('Digite o preço: '))
        vetp.append(p)
    return vetp


def visualizar(vetp):
    if len(vetp) > 0:
        print('Número da Conta    Nome do Titular               Saldo R$')
        for p in vetp:
            print(f'{p.numero_conta:<10} {p.nome:<20} {p.saldo:>10.2f}')
    else:
        print('Não há produtos cadastrados!')


def alterar(vetp):
    if len(vetp) > 0:
        codigo_pesquisa = int(input("Qual o código quer encontrar?"))
        for p in vetp:
            if codigo_pesquisa == p.numero_conta:
                p.nome = input("Digite o novo nome: ")
                p.saldo = float(input("Digite o novo saldo: "))
                break
    else:
        print("Você não tem produtos cadastrados!")
    return vetp

def consultar(vetp):
        if len(vetp) > 0:
            nome_pesquisar = input('Qual nome quer pesquisar? ')
            achei = False
        for p in vetp:
            if nome_pesquisar == p.nome:
                achei = True
                print(f'{p.numero_conta:10} {p.nome:20} {p.saldo:>.2f}')
            if not achei:
            print('Aluno não encontrado!')
    else:
        print('Não há produtos cadastrados!')

def excluir(vetp):
    if len(vetp) > 0:
        for in range(len(vetp)):
            if i == 0:
                menor_saldo = vetp[i].saldo
                indice = i
            if vetp[i].saldo <= menor_saldo:
                menor_saldo = vetp[i].saldo
                indice = i
        vetp.pop(indice)
        print("Exclusão realizada com sucesso")
    else:
        print("Você não tem produtos cadastrados!")

def main():
    vet_cliente = []
    opcao = menu()
    while opcao >= 1 and opcao <= 6:
        if opcao == 1:
            vet_cliente = cadastrar(vet_cliente)
        elif opcao == 2:
            visualizar(vet_cliente)
        elif opcao == 3:
            vet_cliente = consulta(vet_cliente)
        elif opcao == 5:
            vet_cliente = alterar(vet_cliente)
        elif opcao == 6:
            vet_cliente = excluir(vet_cliente)
        opcao = menu()

main()
```

```python
# Por onde começo a programar esse exercício?
# Sempre comece da parte mais simples/talvez a menor e já
# vai implementando e testando cada uma das funções
# Programe o cadastro e o relatório para testar e assim até a
# última função.
class TipoCliente:
    cod_cli = 0
    nome = ''
    fone = ''

class TipoDocumento:
    num_doc = 0
    cod_cli = 0
    dia_venc = 0
    dia_pag = 0
    valor = 0.0
    juros = 0.0


def menu():
    print('\n*** SISTEMA GERENCIADOR DE CLIENTES E DOCUMENTOS ***')
    print('1 - Cadastrar clientes')
    print('2 - Relatório de clientes')
    print('3 - Cadastrar documentos')
    print('4 - Relatório de documentos')
    print('5 - Sair')
    return int(input('Qual opção deseja? '))


def verificar_cliente(codigo, vet_clientes):
    for cliente in vet_clientes:
        if cliente.cod_cli == codigo:
            return True
    return False


def cadastrar_cliente(vet_clientes):
    if len(vet_clientes) >= 15:
        print("Limite de clientes atingido!")
        return

    codigo = int(input("Código do cliente: "))

    if verificar_cliente(codigo, vet_clientes):
        print("Cliente já cadastrado!")
        return

    c = TipoCliente()
    c.cod_cli = codigo
    c.nome = input("Digite o nome do cliente")
    c.fone = input("Digite o telefone do cliente")
    vet_clientes.append(c)
    print("Cliente cadastrado com sucesso!")


def relatorio_clientes(vet_clientes):
    print("\nRELATÓRIO DE CLIENTES")

    for cliente in vet_clientes:
        print(cliente.cod_cli,
              cliente.nome,
              cliente.fone)


def cadastrar_documento(vet_documentos, vet_clientes):
    if len(vet_documentos) >= 30:
        print("Limite de documentos atingido!")
        return
    codigo_cliente = int(input("Código do cliente: "))

    if not verificar_cliente(codigo_cliente, vet_clientes):
        print("Cliente não encontrado!")
        return
    d = TipoDocumento()

    d.num_doc = int(input("Número do documento: "))
    d.cod_cli = codigo_cliente
    d.dia_venc = int(input("Dia do vencimento: "))
    d.dia_pag = int(input("Dia do pagamento: "))
    d.valor = float(input("Valor: "))

    if d.dia_pag > d.dia_venc:
        d.juros = d.valor * 0.05
    else:
        d.juros = 0

    vet_documentos.append(d)
    print("Documento cadastrado com sucesso!")


def relatorio_documentos(vet_documentos):
    print("\nRELATÓRIO DE DOCUMENTOS")

    for doc in vet_documentos:
        print(
            doc.num_doc,
            doc.cod_cli,
            doc.dia_venc,
            doc.dia_pag,
            doc.valor,
            doc.juros
        )


def main():
    vet_clientes = []
    vet_documentos = []
    opcao = menu()
    while opcao >= 1 and opcao <= 4:
        if opcao == 1:
            cadastrar_cliente(vet_clientes)
        elif opcao == 2:
            relatorio_clientes(vet_clientes)
        elif opcao == 3:
            cadastrar_documento(vet_documentos, vet_clientes)
        elif opcao == 4:
            relatorio_documentos(vet_documentos)
        opcao = menu()

main()
```
